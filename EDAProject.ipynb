```python
df_sector=pd.DataFrame(df['activity_description'].groupby
                              (df['activity_description'].iloc[:]).count())
df_sector.rename(columns={"activity_description":"count"},inplace=True)
df_sector.sort_values('count',ascending=False,inplace=True)
df_sector.loc['Others'] = [ df_sector.iloc[5:,0].sum()] 
df_sector.sort_values('count',ascending=False,inplace=True)
df_sector=df_sector.iloc[:6,0]
```

<h5>Now we see how many company registered in a day in months</h5>

```python
df_date=pd.DataFrame(df['date_of_registration'].groupby
                              (df['date_of_registration'].iloc[:]).count())
df_date.rename(columns={"date_of_registration":"count"},inplace=True)
df_date.sort_values('date_of_registration')
df_date.reset_index(inplace=True)
```

```python
month_date=pd.DataFrame(df['date_of_registration'].value_counts())
month_date.reset_index(inplace=True)
month_date['month']='month'
month_date.sort_values('index',inplace=True)

def assign_month(df):
    for i in range(0,len(df.index)):
        if '/01' in df.loc[i,'index']:
            df.loc[i,'month']='Jan-21'
        elif '/02' in df.loc[i,'index']:
            df.loc[i,'month']='Feb-21'
        elif '/03' in df.loc[i,'index']:
            df.loc[i,'month']='Mar-21'
        elif '/04' in df.loc[i,'index']:
            df.loc[i,'month']='Apr-21'
            
assign_month(month_date)
```

```python
ax=plt.figure(figsize=(10,8)).add_axes([0,0,1,1])
sns.set_style('white')
sns.boxplot(x='month',y='date_of_registration',data=month_date)
plt.title('Registered Companies in a day',fontsize=14, weight='bold')
plt.xlabel('Month',fontsize=14)
plt.ylabel('Number of Companies Registered in a day',fontsize=14)
plt.show()
```
